```python
import kagglehub
import pandas as pd
import os
import matplotlib.pyplot as plt

data_path = kagglehub.dataset_download("senadu34/air-quality-index-in-jakarta-2010-2021")
df = pd.read_csv(data_path + '/ispu_dki1.csv')

df['tanggal'] = pd.to_datetime(df['tanggal'])
df['tahun'] = df['tanggal'].dt.year

pollutants = ['pm25', 'pm10', 'so2', 'co', 'o3', 'no2']
yearly_avg = df.groupby('tahun')[pollutants].mean().reset_index()

yearly_avg.to_json('poll_data.json', orient="records")
```
